```python
import urllib.request
from bs4 import BeautifulSoup


names=[]
ratings=[]
year=[]
votes=[]
gross=[]
metascore=[]

years=[str(i) for i in range(2000,2019)]


for yr in years:
    soup=BeautifulSoup(urllib.request.urlopen("https://www.imdb.com/search/title/?release_date="+yr+"&sort=num_votes,desc").read())
    body=soup.find_all('div',class_='lister-item mode-advanced')
    
    for i in body:
        if i.find('div',class_='inline-block ratings-metascore') is not None:
            txt=i.h3.a.text
            names.append(txt)
        
            rat=i.strong.text
            ratings.append(rat)
        
            yr=i.h3.find('span',class_='lister-item-year text-muted unbold').text
            year.append(yr)
        
            vt=i.find('span',attrs={'name':'nv'})['data-value']
            votes.append(vt)
        
            gr=i.find('p',class_="sort-num_votes-visible").text
            gr=gr.split("Gross:\n")[-1]
            gr=gr.split("\n")[0]
            gross.append(gr)
        
            met=i.find('span',class_="metascore").text
            metascore.append(met)
```

```python
import pandas as pd
movie_ratings = pd.DataFrame({'movie': names,
                              'year': year,
                              'imdb': ratings,
                              'metascore': metascore,
                              'votes': votes,
                              'gross': gross})
```
